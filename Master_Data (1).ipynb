**Creating the demographics master data for this (2020,2021,2022,2023)years to know the cases and death rates.**


```python
import pandas as pd

# Load the CSV file into a DataFrame
data = pd.read_csv('drive/MyDrive/chicago_health_disparity/Chicago_Health_disparity/Chicago_covid_daily.csv')

# Convert the 'Date' column to datetime format with dayfirst=True
data['Date'] = pd.to_datetime(data['Date'], format='%m/%d/%Y', dayfirst=True)


# Select the cases columns of interest
columns_of_interest = ['Cases - Latinx', 'Cases - Asian Non-Latinx', 'Cases - White Non-Latinx',
                       'Cases - Black Non-Latinx', 'Cases - Other Race Non-Latinx', 'Cases - Unknown Race/Ethnicity']

# Select the death columns of interest
columns_of_death_interest = ['Deaths - Latinx', 'Deaths - Asian Non-Latinx', 'Deaths - White Non-Latinx',
                       'Deaths - Black Non-Latinx', 'Deaths - Other Race Non-Latinx', 'Deaths - Unknown Race/Ethnicity']

# Filter the data for the year 2020
data_2020 = data[data['Date'].dt.year == 2020]


# Create a new DataFrame with the selected case columns for the year 2020
filtered_data_2020 = data_2020[columns_of_interest]

# Create a new DataFrame with the selected death columns for the year 2020
death_filtered_data_2020 = data_2020[columns_of_death_interest]

#///////////////////////////////////////////////////////////////////////////////////

# Filter the data for the year 2021
data_2021 = data[data['Date'].dt.year == 2021]

# Create a new DataFrame with the selected case columns for the year 2021
filtered_data_2021 = data_2021[columns_of_interest]

# Create a new DataFrame with the selected death columns for the year 2021
death_filtered_data_2021 = data_2021[columns_of_death_interest]

#///////////////////////////////////////////////////////////////////////////////////

# Filter the data for the year 2022
data_2022 = data[data['Date'].dt.year == 2022]


# Create a new DataFrame with the selected case columns for the year 2022
filtered_data_2022 = data_2022[columns_of_interest]

# Create a new DataFrame with the selected death columns for the year 2022
death_filtered_data_2022 = data_2022[columns_of_death_interest]


#///////////////////////////////////////////////////////////////////////////////////

# Filter the data for the year 2023
data_2023 = data[data['Date'].dt.year == 2023]

# Create a new DataFrame with the selected case columns for the year 2023
filtered_data_2023 = data_2023[columns_of_interest]

# Create a new DataFrame with the selected death columns for the year 2023
death_filtered_data_2023 = data_2023[columns_of_death_interest]

# ///////////////////////////////////////////////////////////////////////////////////


#Calculating the cases and deaths rates of 2020#

# Calculate the total cases for each racial category
total_cases_by_race_2020 = filtered_data_2020.sum()
total_deaths_by_race_2020 = death_filtered_data_2020.sum()

# Chicago population census value of 2020
chicago_population_2020 = 2746388

case_rate_per_population_by_race_2020 = total_cases_by_race_2020 / chicago_population_2020

# Calculate the death rate per population for each race
death_rate_per_population_by_race_2020 = total_deaths_by_race_2020 / chicago_population_2020

print("Cases and deaths rates of 2020")

# Print the total cases by race
print(case_rate_per_population_by_race_2020)

# Print the total deaths by race
print(death_rate_per_population_by_race_2020)

#///////////////////////////////////////////////////////////////////////////////////////////////

print("Cases and deaths rates of 2021")
#Calculating the cases and deaths rates of 2021#

total_cases_by_race_2021 = filtered_data_2021.sum()
total_deaths_by_race_2021 = death_filtered_data_2021.sum()

# Chicago population census value of 2021
chicago_population_2021 = 2697000

case_rate_per_population_by_race_2021 = total_cases_by_race_2021 / chicago_population_2021

# Calculate the death rate per population for each race
death_rate_per_population_by_race_2021 = total_deaths_by_race_2021 / chicago_population_2021


# Print the total cases by race
print(case_rate_per_population_by_race_2021)

# Print the total deaths by race
print(death_rate_per_population_by_race_2021)

#///////////////////////////////////////////////////////////////////////////////////////////////

print("Cases and deaths rates of 2022")

#Calculating the cases and deaths rates of 2022#


total_cases_by_race_2022 = filtered_data_2022.sum()
total_deaths_by_race_2022 = death_filtered_data_2022.sum()

# Chicago population census value of 2022
chicago_population_2022 = 2701000

case_rate_per_population_by_race_2022 = total_cases_by_race_2022 / chicago_population_2022

# Calculate the death rate per population for each race
death_rate_per_population_by_race_2022 = total_deaths_by_race_2022 / chicago_population_2022


# Print the total cases by race
print(case_rate_per_population_by_race_2022)

# Print the total deaths by race
print(death_rate_per_population_by_race_2022)

#///////////////////////////////////////////////////////////////////////////////////////////////

print("Cases and deaths rates of 2023")

#Calculating the cases and deaths rates for 2023#


total_cases_by_race_2023 = filtered_data_2023.sum()
total_deaths_by_race_2023 = death_filtered_data_2023.sum()

# Chicago population census value of 2023
chicago_population_2023 = 2704000

case_rate_per_population_by_race_2023 = total_cases_by_race_2023 / chicago_population_2023

# Calculate the death rate per population for each race
death_rate_per_population_by_race_2023 = total_deaths_by_race_2023 / chicago_population_2023


# Print the total cases by race
print(case_rate_per_population_by_race_2023)

# Print the total deaths by race
print(death_rate_per_population_by_race_2023)

#///////////////////////////////////////////////////////

# Calculate case rate change before and after 2022
case_rate_change_before_2022 = case_rate_per_population_by_race_2021 - case_rate_per_population_by_race_2020
case_rate_change_after_2022 = case_rate_per_population_by_race_2023 - case_rate_per_population_by_race_2022

# Print the rate changes
print("Rate Change before 2022:", case_rate_change_before_2022)
print("Rate Change after 2022:", case_rate_change_after_2022)

#///////////////////////////////////////////////////////

# Calculate death rate change before and after 2022
death_rate_change_before_2022 = death_rate_per_population_by_race_2021 - death_rate_per_population_by_race_2020
death_rate_change_after_2022 = death_rate_per_population_by_race_2023 - death_rate_per_population_by_race_2022

# Print the rate changes
print("Rate Change before 2022:", death_rate_change_before_2022)
print("Rate Change after 2022:", death_rate_change_after_2022)
```

**census tract wise covid and death cases.**

```python
zip_data= pd.read_csv('COVID-19_Cases__Tests__and_Deaths_by_ZIP_Code.csv')

# Convert 'Week Start' and 'Week End' columns to datetime
zip_data['Week Start'] = pd.to_datetime(zip_data['Week Start'])
zip_data['Week End'] = pd.to_datetime(zip_data['Week End'])

# Filter the zip code data for the (2020,2021,2022,2023)
data_2020 = zip_data[(zip_data['Week Start'].dt.year == 2020)]
data_2021 = zip_data[(zip_data['Week Start'].dt.year == 2021)]
data_2022 = zip_data[(zip_data['Week Start'].dt.year == 2022)]
data_2023 = zip_data[(zip_data['Week Start'].dt.year == 2023)]


# Group the data by 'ZIP Code', 'Week Start', and 'Week End', and calculate aggregates
grouped_data_2020 = data_2020.groupby(['ZIP Code']).agg({
    'Cases - Weekly': 'sum',
    'Deaths - Weekly': 'sum',
}).reset_index()

grouped_data_2021 = data_2021.groupby(['ZIP Code']).agg({
    'Cases - Weekly': 'sum',
    'Deaths - Weekly': 'sum',
}).reset_index()

grouped_data_2022 = data_2022.groupby(['ZIP Code']).agg({
    'Cases - Weekly': 'sum',
    'Deaths - Weekly': 'sum',
}).reset_index()

grouped_data_2023 = data_2023.groupby(['ZIP Code']).agg({
    'Cases - Weekly': 'sum',
    'Deaths - Weekly': 'sum',
}).reset_index()

# Read the census tract file of all the years
census_tract_data_2020 = pd.read_csv('drive/MyDrive/chicago_health_disparity/Chicago_Health_disparity/hud_census_tract_2020.csv')

census_tract_data_2021 = pd.read_csv('drive/MyDrive/chicago_health_disparity/Chicago_Health_disparity/hud_census_tract_2021.csv')

census_tract_data_2022 = pd.read_csv('drive/MyDrive/chicago_health_disparity/Chicago_Health_disparity/hud_census_tract_2022.csv')

census_tract_data_2023 = pd.read_csv('drive/MyDrive/chicago_health_disparity/Chicago_Health_disparity/hud_census_tract_2023.csv')

census_tract_data_2020['ZIP'] = census_tract_data_2020['ZIP'].astype(str)
census_tract_data_2021['ZIP'] = census_tract_data_2021['ZIP'].astype(str)
census_tract_data_2022['ZIP'] = census_tract_data_2022['ZIP'].astype(str)
census_tract_data_2023['ZIP'] = census_tract_data_2023['ZIP'].astype(str)

# Merge ZIP code and census tract data
merged_data_2020 = grouped_data_2020.merge(census_tract_data_2020, left_on='ZIP Code', right_on='ZIP', how='inner')

merged_data_2021 = grouped_data_2021.merge(census_tract_data_2021, left_on='ZIP Code', right_on='ZIP', how='inner')

merged_data_2022 = grouped_data_2022.merge(census_tract_data_2022, left_on='ZIP Code', right_on='ZIP', how='inner')

merged_data_2023 = grouped_data_2023.merge(census_tract_data_2023, left_on='ZIP Code', right_on='ZIP', how='inner')


# Calculate estimated cases per census tract 2020
merged_data_2020['Estimated_Cases_Tract_2020'] = merged_data_2020['Cases - Weekly'] * merged_data_2020['RES_RATIO']

merged_data_2020['Estimated_Deaths_Tract_2020'] = merged_data_2020['Deaths - Weekly'] * merged_data_2020['RES_RATIO']

# Calculate estimated cases per census tract 2021
merged_data_2021['Estimated_Cases_Tract_2021'] = merged_data_2021['Cases - Weekly'] * merged_data_2021['RES_RATIO']

merged_data_2021['Estimated_Deaths_Tract_2021'] = merged_data_2021['Deaths - Weekly'] * merged_data_2021['RES_RATIO']

# Calculate estimated cases per census tract 2022
merged_data_2022['Estimated_Cases_Tract_2022'] = merged_data_2022['Cases - Weekly'] * merged_data_2022['RES_RATIO']

merged_data_2022['Estimated_Deaths_Tract_2022'] = merged_data_2022['Deaths - Weekly'] * merged_data_2022['RES_RATIO']

# Calculate estimated cases per census tract 2023
merged_data_2023['Estimated_Cases_Tract_2023'] = merged_data_2023['Cases - Weekly'] * merged_data_2023['RES_RATIO']

merged_data_2023['Estimated_Deaths_Tract_2023'] = merged_data_2023['Deaths - Weekly'] * merged_data_2023['RES_RATIO']

# #Remove all the unnecessary data from

final_merged_data_2020 = merged_data_2020[['ZIP Code','TRACT','Estimated_Cases_Tract_2020','Estimated_Deaths_Tract_2020']]
final_merged_data_2021 = merged_data_2021[['ZIP Code','TRACT','Estimated_Cases_Tract_2021','Estimated_Deaths_Tract_2021']]
final_merged_data_2022 = merged_data_2022[['ZIP Code','TRACT','Estimated_Cases_Tract_2022','Estimated_Deaths_Tract_2022']]
final_merged_data_2023 = merged_data_2023[['ZIP Code','TRACT','Estimated_Cases_Tract_2023','Estimated_Deaths_Tract_2023']]


# Save the final data to of 2020 to new CSV file
final_merged_data_2020.to_csv('census_tract_case_rates_2020.csv', index=False)

# Save the final data of 2020 to a new CSV file
final_merged_data_2021.to_csv('census_tract_case_rates_2021.csv', index=False)

# Save the final data of 2020 to a new CSV file
final_merged_data_2022.to_csv('census_tract_case_rates_2022.csv', index=False)

# Save the final data of 2020 to a new CSV file
final_merged_data_2023.to_csv('census_tract_case_rates_2023.csv', index=False)
```
